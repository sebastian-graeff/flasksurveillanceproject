```python
from geopy import geocoders  
gn = geocoders.GeoNames(username="sebasg")

#extract latitude and longitude from output
hq_latitude = []
hq_longitude = []

for hq_city in companies_df.hq_city: 
    try:
        hq_city = gn.geocode(hq_city, exactly_one=False)[0]
        hq_latitude_val = hq_city.latitude
        hq_longitude_val = hq_city.longitude
    except TypeError:
        hq_latitude_val = np.nan
        hq_longitude_val = np.nan

    hq_latitude.append(hq_latitude_val)
    hq_longitude.append(hq_longitude_val)
```

```python
#make new series in dataframe
companies_df["hq_latitude"] = hq_latitude
companies_df["hq_longitude"] = hq_longitude
```

Now calculate closest hq to a user given location

```python
def closest_surveillance_company(user_input_name):

    list_of_coords = list(zip(companies_df.hq_latitude, companies_df.hq_longitude))
    user_city = gn.geocode(user_input_name, exactly_one=False)[0]
    user_city_coord = (user_city.latitude, user_city.longitude)

    distance_dict = dict()

    for position,coord in enumerate(list_of_coords):
        try:
            coord_distance = geopy.distance.distance(user_city_coord, coord).km
            distance_dict[position] = coord_distance
        except ValueError:
            pass
    
    index_lowest = min(distance_dict, key=distance_dict.get)
    entry = companies_df.iloc[index_lowest]
    return entry.company_name
```

```python
#get min distance location
closest_surveillance_company("London")
```
